```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
B0 = 5e-9
T = 1 * 60
sigma = 3.3
mu_0 = 4 * np.pi * 1e-7
omega = 2 * np.pi / T

# Depth range
depth = np.linspace(0, 20000, 100)

# Time instances
time = [0, 7.5, 15, 30, 45]

# skin depth
def skin_depth(omega, sigma, mu):
    return np.sqrt(2 / (omega * mu * sigma))

# calculating magnetic field and electric field
def magneticelectric_field(B0, omega, depth, time):
    zs = skin_depth(omega, sigma, mu_0)
    print('Skin depth:',zs)
    z = depth
    Bx = B0 * np.exp(1j * omega * time) * np.exp(-(1 + 1j) * z / zs)
    By = 0
    Ex = (1 + 1j) * By / (zs* mu_0 * sigma)
    Ey = (-1 + 1j) * Bx / (zs * mu_0 * sigma)

    return Bx, By, Ex, Ey



# Plotting only the magnetic field
for i, time in enumerate(time):
    Bx, By, Ex, Ey = magneticelectric_field(B0, omega, depth, time)

    plt.figure()

    plt.plot(np.real(Bx), depth, label='Real part')
    plt.plot(np.imag(Bx), depth, label='Imaginary part')
    plt.title(f'Magnetic Field at t = {time} s')
    plt.xlabel('Magnetic Field (Tesla)')
    plt.ylabel('Depth (m)')
    plt.gca().invert_yaxis()
    plt.legend()




plt.tight_layout()
plt.show()
```
