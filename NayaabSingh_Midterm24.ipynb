```python
## had to do this for my own reference as could not figure something out with the liusts##
time1 = time[:,0]
time2 = time[:,1]
time3 = time[:,2]
time4 = time[:,3]

temp1 = temp[:,0]
temp2 = temp[:,1]
temp3 = temp[:,2]
temp4 = temp[:,3]


plt.plot(time1, temp1, 'dr', markersize = 2)
plt.plot(time2, temp2, 'db', markersize = 2)
plt.plot(time3, temp3, 'dm', markersize = 2)
plt.plot(time4, temp4, 'dc', markersize = 2)

```

```python
def tempchange(time, Af, Ai, t0, tau):
    '''
    Input:
    time - time array
    Af = final temp
    Ai = initial temp
    t0 = threshold time
    tau = exponential factor

    Output:

    A = temp array
    '''
    A = np.where(time<t0, Ai, Af + (Ai-Af)*np.exp(-(time-t0)/tau)) # if less than threshold Ai
                                                                # if more our beloved expression

    return A


###testing values###
Ai = 20
Af = 10
t0 = 40
tau =10

test_time = np.linspace(1,120,1000) ## smooth time

A = tempchange( test_time, Af, Ai, t0, tau) # smooth temp

plt.plot(test_time, A, '-k', label = 'smooth test')

plt.xlabel('time(sec)')
plt.ylabel('temp(deg C)')
plt.title('Test plot for Temp Change function')
plt.legend()

# plotting looks good
```

```python
# manually for temp
temp_4 =[]
print(temp4.shape)
temp_4 = np.append(temp_4, temp1, axis = 0)
temp_4 = np.append(temp_4, temp2, axis = 0)
temp_4 = np.append(temp_4, temp3, axis = 0)
temp_4 = np.append(temp_4, temp4, axis = 0)

print(temp_4)
print(temp_4.shape)
```

```python
plt.plot(time_4, temp_4, '+') # this also works quick check
```

```python
guess=[10, 7.5 , 40 , 10] # guess values

pout, cout = sc.curve_fit(tempchange, time_4, temp_4) # curve fitting
print(pout)

# plotting it out

plt.plot(time_4, temp_4, '+', label ='Raw Data')
smoothtime = np.linspace(min(time_4), max(time_4), 1000)
plt.plot(smoothtime, tempchange(smoothtime,*pout), '-k', label = 'Best fit')
plt.xlabel('time(sec)')
plt.ylabel('temp(deg C)')
plt.title('Best fit(with all data)')
plt.legend()
```
